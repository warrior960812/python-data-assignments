```python
import requests
import csv
import bs4
```

```python
r = requests.get('https://apnews.com/apf-topnews')
```

```python
r.encoding = 'utf-8'
```

```python
r.text
```

```python
mypage=bs4.BeautifulSoup(r.text)
```

```python
mypage
```

```python
Days=mypage.find_all("span",attrs={"class":"Timestamp"})
Days
```

```python
days=[]
for day in Days:
    d=day.text
    days.append(d)
```

```python
days
```

```python
mytitle=mypage.find_all("a",attrs={"class":"headline"})
mytitle
```

```python
titles=[]
for title in mytitle:
    title.find("h1")
    t=title.text
    titles.append(t)
```

```python
titles
```

```python
len(titles)
```

```python
myplace=mypage.find_all("div",attrs={"class":"content"})
places=[]
for place in myplace:
    print(place.find("p"))
    location = place.find("p").text.split('(')[0]  
    places.append(location)
```

```python
places
```

```python
myplace[0].find('p').text.split('(')[0]
```

```python
len(places)
```

```python
byline=mypage.find_all("span",attrs={"class":"byline"})
```

```python
bylines=[]
for Byline in byline:
    n=Byline.text.strip()
    bylines.append(n)
    
```

```python
bylines
```

```python
len(bylines)
```

```python
tag=mypage.find_all("a",attrs={"class":"HubTag"})
tag
```

```python
tags=[]
for Tag in tag:
    tags.append(Tag.text)
tags
```

```python
len(tags)
```

```python
links=mypage.find_all("a",attrs={"class":"headline"})
urls = []
for link in links:
    url = '{0}{1}'.format('https://apnews.com',link['href'])
    urls.append(url)
```

```python
urls
```

```python
len(titles)
```

```python
len(places)
```

```python
len(bylines)
```

```python
len(tags)
```

```python
Z=[]
for i in range(0,48):
    z=[days[i],titles[i],places[i],bylines[i],tags[i],urls[i]]
    Z.append(z)
```

```python
Z
```

```python
with open('topnews.csv', 'a', newline='') as csvfile:
    spamwriter = csv.writer(csvfile,
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
    header= ["days",'titles','places','bylines',"tags","urls"]
    spamwriter.writerow(header)
    spamwriter.writerows(Z)
```
