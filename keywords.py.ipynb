```python
articles=["trade-wars-news1.txt","trade-wars-news2.txt","trade-wars-news3.txt","trade-wars-news4.txt","trade-wars-news5.txt"]
```

```python
articles=[]
for article in articles:
    f = open(articles,'rb')
    content= f.read()
    all_articles.append(content)
    
all_articles
```

```python
type(all_articles)
```

```python
all_articles[1].split()
```

```python
all_words = []
for i in range(5):

    words = all_articles[i].split()
    for all_word in words:
          all_words.append(all_word)
```

```python
all_words
```

```python
len(all_words)
```

```python
from collections import Counter
all_words_dict = {}
for i in all_words:
    all_words_dict[i] = all_words.count(i)
    
all_words_dict
```

```python
type(all_words_dict)
```

```python
sortedwordlist=sorted(all_words_dict.items(), key=lambda asd:asd[1],reverse=True)
sortedwordlist
```

```python
top15=sortedwordlist[:15]
top15
```

```python
import csv
f=open('wordfrequency.csv','w',newline='')
writer=csv.writer(f)
header=['keyword','frequency']
writer.writerow(header)
writer.writerows(top15)
f.close()
```

```python
with open('wordfrequency.csv','r',encoding='utf8') as f:
	rows = csv.reader(f)
	for row in rows:
		print(row)
```
